```python

y_prediction2 = []
for j in np.arange(df_knn_test.shape[0]):
    distance_row = []
    for k in np.arange(df_knn_train.shape[0]):
        distance_row.append(np.linalg.norm(df_knn_train.iloc[k,:-2]-df_knn_test.iloc[j,:-2],ord = 2))
    df_knn_train['distance_for_row'] = distance_row  
    df_knn_train.sort_values(by='distance_for_row',inplace = True)
    y_pred = df_knn_train.iloc[0:4,-2].mean()
#     print(y_pred)
    y_prediction2.append(y_pred)
    df_knn_train.drop(columns = 'distance_for_row',inplace = True)
```
